```python




print("Welcome to Movie Hub")
print("Please enter the name of the movie")
name = input("Enter your movie name: ")

# Filter DataFrame based on user input
matching_movies = movies_df[movies_df["Title"].str.contains(name, case=False, na=False)]

if not matching_movies.empty:
    print("Matching Movies:")
    display(matching_movies)
else:
    print("Movie not found. Enter something else")
    year = input("Enter the year the movie was released in: ")
    
    # Convert year to integer
    try:
        year = int(year)
        matching_movies_by_year = movies_df[movies_df["Year"] == year]
        
        if not matching_movies_by_year.empty:
            print("Movies released in", year)
            display(matching_movies_by_year)
            #print(tabulate(matching_movies, headers='keys', tablefmt='grid'))
        else:
            print("No movies found for the entered year. See you again!")
            genres = input("Enter the genres of the movie")
            try:
                matching_movie_by_genres = movies_df[movies_df["Genres"].str.contains(genres, case=False, na=False)]
                if not matching_movie_by_genres.empty:
                    print("movies by genres")
                    display(matching_movie_by_genres)
                else:
                    print("NO movie with that generes")
            except ValueError:
                print("Invalid year format. Please enter a number.")

    except valueError:
        print("Thank for visiting")
   
```
